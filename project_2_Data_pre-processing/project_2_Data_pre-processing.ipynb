### Замена типа данных

Так как Series 'days_employed' имеет тип object, то сначала мы переведём его в числовой(float) воспользовавшись методом to_numeric(). Типы значений, принимаемых errors, зависят действия to_numeric при встрече с некорректным значением:

errors='raise' — дефолтное поведение: при встрече с некорректным значением выдаётся ошибка, операция перевода в числа прерывается;

errors='coerce' — некорректные значения принудительно заменяются на NaN;

errors='ignore' — некорректные значения игнорируются, но остаются.

Пример:

transactions['amount'] = pd.to_numeric(transactions['amount'], errors='coerce')

```python
#Выведем ещё раз общие сведения
print(df.info())

#Days employed  - определим тип Float
df['days_employed'] = pd.to_numeric(df['days_employed'], errors='raise')

#Рассмотрим срез по 10 срокам
df.head(10)
```

### Лемматизация

Лемматизация будет необходима для создания дополнительного признака по классификации. Выделим классы:

```python
%%time
from tqdm.notebook import tqdm

#импортирую библеотеку для лематизации
from pymystem3 import Mystem
m = Mystem()

#записываю анализируемый столбец с его уникальными записями
purpose_column = df['purpose'].drop_duplicates()

#блок контейнеров для анализа, многие из них буферные
list_acum=[]
dict_list_acum={}
purpose = []
dict_p = {}
c = []

for row in tqdm(purpose_column):
    lemma = m.lemmatize(row) #лемматизация уникального элемента сделана
    
    c=[] #лемматизация уникального элемента сделана
    
    list_acum.append(lemma) #контейнер для списка лемм-х строк
    dict_list_acum[row]=lemma #словарь для сопоставления значений до и после лемматизации
    
    #для создания лематизированой строки без переноса и записи в единый элемент
    for i in lemma:
        purpose.append(i)
        if (i != '\n') & (i != ' '):
            c.append(i)
    #соединение слов после лемматизации одной строки      
    lemma = ' '.join(c)    
    dict_p[row] = lemma #словарь для сопоставления значений до и после (объединение слов в строку) лемматизации
    print('ДО |{: <40}| ПОСЛЕ |{: <40}|'.format(row, lemma))
```

```python
from collections import Counter

list_purpose = list(Counter(purpose).elements())
list_purpose_new = []

for i in list_purpose:
    if (i != '')&(i != ' ')&(i != ', ')&(i != '\n')&(i != 'со')&(i != 'свой')&(i != 'для')&(i != 'на')&(i != 'с'):
        list_purpose_new.append(i)
Counter(list_purpose_new)  
```

```python
#основные категории целей
short_purpose = ['жилье','автомобиль','образование','свадьба','операция','недвижимость']

dict_list_acum_value = {}

for row in dict_p.values(): #for row in purpose_column
    for item in short_purpose:
        if item in row:
            dict_list_acum_value[row] = item
```

```python
df['purpose_lemma'] = df['purpose']

for item in list(dict_p.keys()):
    df.loc[df['purpose_lemma'] == item, 'purpose_lemma'] = dict_p[item]

df['purpose_lemma_word'] = df['purpose_lemma']

for item in list(dict_list_acum_value.keys()):
    df.loc[df['purpose_lemma'] == item, 'purpose_lemma_word'] = dict_list_acum_value[item]

df.head()
df['purpose_lemma_word'].value_counts()
```

Рассмотри получаемый доход и категоризируем его:

```python
print('минимальная з/п', df.loc[df['total_income'] != 0, 'total_income'].min())
print('максимальная з/п', df['total_income'].max())
print('средняя з/п', df['total_income'].mean())
print('медиана з/п', df['total_income'].median())
```

```python
def func_benefit(row):
    benefit = row['total_income']
    if benefit == 0:
        return 'нет дохода'
    elif 0 < benefit <= 90000:
        return 'низкий'
    elif 90000 < benefit <= 181000:
        return 'средний'
    elif 181000 <= benefit <= 250000:
        return 'высокий'
    else:
        return 'очень высокий'

    
df['total_income_status'] = df.apply(func_benefit, axis = 1)
print(df['total_income_status'].value_counts())
```

Введены несколько категорий, за среднее взято медианное значение ряда, мимнимальный доход до 90000, выше уже среднее значение, макисмальне значение мреднего принято как разница между масимальным низким доходом и среднем значением.

Так как вопрос стоит влияет ли семейное положение и количество детей на факт погашения, то введём следующую классификацию
    0 - нет детей
    больше 0 - есть дети
Вначале лучше проверить гипотезу, что человек у которого есть дети более аккуратен к факту погашения

```python
def func_child(row):
    child = row['children']
    if child != 0:
        return 'есть дети'
    else:
        return 'нет детей'
    
def func_family_new_status(row):
    '''
    если женат / замужем  - в браке
    если гражданский брак - в браке
    если Не женат / не замужем - холост
    если в разводе - холост
    если вдовец / вдова - холост
    '''
    status = row['family_status']
    if status == 'женат / замужем':
        return 'в браке'
    elif status == 'гражданский брак':
        return 'в браке'
    elif status == 'Не женат / не замужем':
        return 'не в браке'
    elif status == 'в разводе - холост':
        return 'не в браке'
    else:
        return 'не в браке'
    
df['children_status'] = df.apply(func_child, axis = 1)
df['family_new_status'] = df.apply(func_family_new_status, axis = 1)
df.head()
```

```python
df['income_type'].value_counts()
```

```python
def func_work_status(row):
    '''
    для скоринга необходимо выделить более общие классы,
    т.е. понять, обратившийся в банк клиент работает или нет,
    то для первоначальной оценки этого достаточно.
    '''
    status = row['family_status']
    
    if status == 'сотрудник':
        return 'работает'
    elif status == 'компаньон':
        return 'работает'
    elif status == 'госслужащий':
        return 'работает'
    elif status == 'предприниматель':
        return 'работает'
    else:
        return 'безработный'
    
df['work_status'] = df.apply(func_work_status, axis=1)
df.head()
```
