## Описание технологического процесса


Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

### 1. Флотация
Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).


### 2. Очистка
Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Описание данных 
1. Технологический процесс
    * Rougher feed — исходное сырье
    * Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    * Xanthate **— ксантогенат (промотер, или активатор флотации);
    * Sulphate — сульфат (на данном производстве сульфид натрия);
    * Depressant — депрессант (силикат натрия).
    * Rougher process (англ. «грубый процесс») — флотация
    * Rougher tails — отвальные хвосты
    * Float banks — флотационная установка
    * Cleaner process — очистка
    * Rougher Au — черновой концентрат золота
    * Final Au — финальный концентрат золота
2. Параметры этапов
    * air amount — объём воздуха
    * fluid levels — уровень жидкости
    * feed size — размер гранул сырья
    * feed rate — скорость подачи


### Наименование признаков
Наименование признаков должно быть такое:
__[этап].[тип_параметра].[название_параметра]__ Пример: rougher.input.feed_ag


Возможные значения для блока [этап]:
* rougher — флотация
* primary_cleaner — первичная очистка
* secondary_cleaner — вторичная очистка
* final — финальные характеристики


Возможные значения для блока [тип_параметра]:
* input — параметры сырья
* output — параметры продукта
* state — параметры, характеризующие текущее состояние этапа
* calculation — расчётные характеристики

Дублей нет во всех сетах. Для дальнейшего анализа - разобьём тестовый файл на части по стадиям тех.процесса

```python
def step_process(data=data_train , date=True, table=True):
    '''
    Подфункция, которая обрабатывает названия признаков и на основе лемматизации сопоставляет признак и стадию тех.процесса.
    На выводе получает списки наименований для последующего построения таблиц. Если указать в параметре функции date=False,
    то созданные списки, а в последующем и таблицы не будут содержать дату. Парметр table в значении True позволяет получить
    в выводе таблицы, а при значении False - не будет выводить таблицы.
    '''
    
    rougher = []
    primary = []
    secondary = []
    final = []

    short_step = ['rougher', 'primary', 'secondary', 'final']
    for col in data.columns:
        if m.lemmatize(col)[0] == short_step[0]:
            rougher.append(col)
        elif m.lemmatize(col)[0] == short_step[1]:
            primary.append(col)
        elif m.lemmatize(col)[0] == short_step[2]:
            secondary.append(col)
        else:
            final.append(col)
            
    if date==True:
        rougher.append('date')
        primary.append('date')
        secondary.append('date')
        final.append('date') 
        
    return rougher, primary, secondary, final

def step_process_in_column(data, date=True, table=True):
    '''
    Функция выдает четыре таблицы соответствующие этапам Тех.процесса по извлечению золота
    '''       
    if table==True:
        rougher_df = data[step_process(data, date)[0]]
        primary_df = data[step_process(data, date)[1]]
        secondary_df = data[step_process(data, date)[2]]
        final_df = data[step_process(data, date)[3]]
        
        return rougher_df, primary_df, secondary_df, final_df
    else:
        return step_process(data , date, table)
       
rougher_df = step_process_in_column(data_train, True, True)[0]
primary_df = step_process_in_column(data_train, True, True)[1]
secondary_df = step_process_in_column(data_train, True, True)[2]
final_df = step_process_in_column(data_train, True, True)[3]

rougher_df.head()
```

## MAPE and sMAPE

где:
* C — доля золота в концентрате после флотации/очистки;
* F — доля золота в сырье/концентрате до флотации/очистки;
* T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

```python
#поиск нужных признаков
for i in list(data_train.columns):
    if (((('concentrate' in m.lemmatize(i)) | ('feed' in m.lemmatize(i)))&('au' in m.lemmatize(i)))\
    | ('recovery' in m.lemmatize(i))) & ('calculation' not in m.lemmatize(i)) :
        print(i)
```

```python
def func_mape(row):
    #доля золота в концентрате после флотации/очистки;
    C = row['rougher.output.concentrate_au']
    #доля золота в сырье/концентрате до флотации/очистки;
    F = row['rougher.input.feed_au']
    #доля золота в отвальных хвостах после флотации/очистки
    T = row['rougher.output.tail_au']
    try:
        return (C*(F-T))/(F*(C-T))*100
    except ZeroDivisionError:
        return 0
    
def func_mape_2(row):
    #доля золота в концентрате после флотации/очистки;
    C = row['final.output.concentrate_au']
    #доля золота в сырье/концентрате до флотации/очистки;
    F = row['rougher.input.feed_au']
    #доля золота в отвальных хвостах после флотации/очистки
    T = row['final.output.tail_au'] 
    try:
        return (C*(F-T))/(F*(C-T))*100
    except ZeroDivisionError:
        return 0

data_train['rougher.output.recovery_2'] = data_train.apply(func_mape, axis=1)
data_train['final.output.recovery_2'] = data_train.apply(func_mape_2, axis=1)
data_train.loc[:,['rougher.output.recovery','rougher.output.recovery_2',\
                  'final.output.recovery', 'final.output.recovery_2']].head()
```

```python
result = pd.DataFrame(columns=['name','type','estimator','depth','mae','r2',\
                               'sMAPE: final recovery','sMAPE: rougher recovery', 'TOTAL sMAPE'])

def new_parametrs(name,types,estimator,depth,y_test, y_predict, counter=True):
    if counter == True:
        i = np.max(result.index)+1
    else:
        i = 0
    result.loc[i] = [name,types,estimator,depth,\
                     mean_absolute_error(y_test, y_predict),\
                     r2_score(y_test, y_predict),\
                     smape(y_test, y_predict)[0][0],\
                     smape(y_test, y_predict)[0][1],\
                     smape(y_test, y_predict)[1]\
                    ]
    return result
```
